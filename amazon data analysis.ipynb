```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("amazon.csv")
```

```python
# Group by product name and count reviews
popular_products = df.groupby("product_name").size().reset_index(name="count")
popular_products = popular_products.sort_values("count", ascending=False).head(10)

# Show the DataFrame
print(popular_products)

# Limit product_name to a maximum of 3 words
popular_products['product_name'] = popular_products['product_name'].apply(lambda x: ' '.join(x.split()[:3]))

# Bar chart
plt.figure(figsize=(4,4))
sns.barplot(x="product_name", y="count", data=popular_products, palette="viridis")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.xlabel("Product Name", fontsize=10)
plt.ylabel("Reviews", fontsize=10)
plt.title("Top 10 Popular Products", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Aggregate: Average rating and review count per category
category_stats = df.groupby("category").agg(
    avg_rating=("rating", "mean"),
    total_reviews=("review_id", "count")
).reset_index().sort_values("total_reviews", ascending=False)

print(category_stats.head(5))

# Filter top 5 categories based on total reviews
top_5_categories = category_stats.head(5)

top_5_categories['category'] = top_5_categories['category'].apply(lambda x: ' '.join(x.split('|')[:3]))

# Dual-axis chart (using matplotlib for secondary y-axis)
fig, ax1 = plt.subplots(figsize=(8, 4)) 

# Bar chart for total reviews
sns.barplot(x="category", y="total_reviews", data=top_5_categories, ax=ax1, palette="Blues_d")
ax1.set_ylabel("Total Reviews", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
plt.xticks(rotation=60, ha='right', fontsize=8)  

# Create a secondary y-axis for average rating
ax2 = ax1.twinx()
sns.lineplot(x="category", y="avg_rating", data=top_5_categories, ax=ax2, sort=False, marker="o", color="red")
ax2.set_ylabel("Average Rating", color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.xlabel("Category", fontsize=8)  
plt.title("Top 5 Categories: Average Rating and Total Reviews", fontsize=10) 
plt.tight_layout()
plt.show()
```

```python
# Calculate the overall average rating
avg_rating = df["rating"].mean()

# Filter products with discount > 50% and rating above overall average
high_discount_good_rating = df[(df["discount_percentage"] > 50) & (df["rating"] > avg_rating)]
high_discount_good_rating = high_discount_good_rating[["product_name", "discount_percentage", "rating"]].drop_duplicates()
high_discount_good_rating = high_discount_good_rating.sort_values("discount_percentage", ascending=False)

print(high_discount_good_rating.head(10))
# Filter top 10 products with the highest discount and above-average rating
top_10_high_discount = high_discount_good_rating.head(10)

top_10_high_discount['product_name'] = top_10_high_discount['product_name'].apply(lambda x: ' '.join(x.split()[:3]))

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_10_high_discount, x="discount_percentage", y="rating", hue="product_name", palette="deep", s=100)
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.title("Top 10 Products with High Discounts and Above-Average Ratings")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Group by category to get average discount percentage
discount_distribution = df.groupby("category")["discount_percentage"].mean().reset_index()
discount_distribution = discount_distribution.sort_values("discount_percentage", ascending=False)

print(discount_distribution.head(8))
# Get top 8 categories with the highest average discount percentage
top_8_discount_distribution = discount_distribution.head(8)

top_8_discount_distribution['category'] = top_8_discount_distribution['category'].apply(lambda x: ' '.join(x.split()[:4]))

# Plot: Horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x="discount_percentage", y="category", data=top_8_discount_distribution, palette="coolwarm")
plt.xlabel("Average Discount Percentage")
plt.ylabel("Category")
plt.title("Top 8 Categories: Distribution of Average Discounts")
plt.tight_layout()
plt.show()
```

```python
# Select required columns
rating_discount = df[["rating", "discount_percentage"]]
print(rating_discount.head())

# Scatter plot with regression line
plt.figure(figsize=(10,6))
sns.regplot(x="discount_percentage", y="rating", data=rating_discount, scatter_kws={'alpha':0.5})
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.title("Correlation between Discount Percentage and Rating")
plt.tight_layout()
plt.show()
```

```python
# Group by user_name to count reviews
top_reviewers = df.groupby("user_name").size().reset_index(name="count")
top_reviewers = top_reviewers.sort_values("count", ascending=False).head(5)
print(top_reviewers)

# Bar chart
plt.figure(figsize=(10,6))
sns.barplot(x="user_name", y="count", data=top_reviewers, palette="magma")
plt.xlabel("User Name")
plt.ylabel("Number of Reviews")
plt.title("Top 5 Users by Number of Reviews")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Aggregate total sales per category (using discounted_price as proxy)
category_sales = df.groupby("category")["discounted_price"].sum().reset_index()
category_sales = category_sales.sort_values("discounted_price", ascending=False).head(5)

print(category_sales)

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(category_sales["discounted_price"], labels=category_sales["category"], autopct='%1.1f%%', startangle=140)
plt.title("Category-wise Total Sales (Discounted Price Sum)")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Filter products with rating below 3 and discount greater than 50%
low_rating_high_discount = df[(df["rating"] < 3) & (df["discount_percentage"] > 50)]
low_rating_high_discount = low_rating_high_discount[["product_name", "rating", "discount_percentage"]].drop_duplicates()
low_rating_high_discount = low_rating_high_discount.sort_values("discount_percentage", ascending=False)

print(low_rating_high_discount)

# Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=low_rating_high_discount, x="rating", y="discount_percentage", hue="product_name", palette="Set2", s=100)
plt.xlabel("Rating")
plt.ylabel("Discount Percentage")
plt.title("Low Rated Products with High Discounts")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

# Identify top categories by review count
top_categories = df.groupby("category").size().reset_index(name="count").sort_values("count", ascending=False).head(5)["category"].tolist()

for category_name in top_categories:
    # Collect all review titles for the category
    category_reviews = df.loc[df["category"] == category_name, "review_title"].dropna().tolist()
    text = " ".join(category_reviews)
    
    # Generate WordCloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Plot the WordCloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"WordCloud for Review Titles in Category: {category_name}")
    plt.tight_layout()
    plt.show()
```
